```python
statement_counts = df.groupby('customer_ID').size()

# Calculate min, max, and average
min_statements = statement_counts.min()
max_statements = statement_counts.max()
avg_statements = statement_counts.mean()
median_statements = statement_counts.median()

print(f"\nMinimum number of statements per customer: {min_statements}")
print(f"Maximum number of statements per customer: {max_statements}")
print(f"Average number of statements per customer: {avg_statements:.2f}")
print(f"Median number of statements per customer: {median_statements:.2f}")
```

```python
total_null_percentage = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
print(f"Total null percentage in the dataset: {total_null_percentage:.2f}%")
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
highest_missing_features = missing_percentage.sort_values(ascending=False).head(10)
print("Features with the highest percentage of missing values:")
print(highest_missing_features)
```

```python
print(highest_missing_features.index.tolist())
```

```python
categorical_cols = [
    'B_30', 'B_38', 'D_63', 'D_64', 'D_66', 'D_68',
    'D_114', 'D_116', 'D_117', 'D_120', 'D_126'
]

for col in categorical_cols:
    if col in df.columns:
        print(f"\n--- Value counts for: {col} ---")
        try:
            # Use value_counts to see the distribution of each category, including NaNs
            counts = df[col].value_counts(dropna=False).reset_index()
            counts.columns = ['Value', 'Count'] # Rename columns for clarity
            # Calculate percentage
            counts['Percentage'] = (counts['Count'] / df.shape[0]) * 100
            display(counts)
        except Exception as e:
            print(f"Could not get value counts for {col}: {e}")
    else:
        print(f"\nNote: Categorical column '{col}' not found in the loaded training dataframe.")

```

```python
df['S_2'] = df['S_2'].astype('datetime64[ns]')
df['end_of_month'] = df['S_2'] + pd.offsets.MonthEnd(0)
```

```python
targets = df[df['end_of_month'] == '2018-03-31'][['customer_ID', 'target']]
```

```python
summary_stats_labels = targets.describe(include='all').T
display(summary_stats_labels)
```

```python
targets.isna().sum()
```

```python
target_counts = targets['target'].value_counts(dropna=False)
# Calculate percentages
target_percentages = targets['target'].value_counts(normalize=True, dropna=False) * 100
# Combine into a single DataFrame for nice display
target_dist = pd.DataFrame({'Count': target_counts, 'Percentage': target_percentages})
display(target_dist)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

non_feature_cols = ['customer_ID', 'S_2', 'end_of_month', 'target']
feature_cols = [col for col in df.columns if col not in non_feature_cols]
# Calculate missing percentage for each feature column, grouped by statement date
# This creates a DataFrame where index is S_2, columns are features, values are % missing
missing_over_time = df.groupby('end_of_month')[feature_cols].apply(lambda x: x.isnull().mean() * 100)

# Calculate the average missing percentage across ALL features for each date
overall_missing_per_date = missing_over_time.mean(axis=1)

# --- Plotting ---
plt.figure(figsize=(15, 6))
sns.lineplot(x=overall_missing_per_date.index, y=overall_missing_per_date.values)
plt.title('Overall Percentage of Missing Data per Statement Date (S_2)')
plt.xlabel('Statement Date (S_2)')
plt.ylabel('Average Missing Data (%) Across All Features')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("Analysis Complete. Observe the plot above for trends.")
# Interpretation Hint: Look for upward/downward trends or specific periods with spikes.
```

```python
# Use the same selected features as before, or choose others
selected_features = ['P_2', 'D_39', 'B_1', 'S_3', 'R_1']

# Check if selected features exist in the DataFrame
missing_selection = [f for f in selected_features if f not in df.columns]
if missing_selection:
    print(f"Warning: The following selected features are missing: {missing_selection}")
    selected_features = [f for f in selected_features if f in df.columns]

if selected_features:
    print("\nGenerating Histograms and Box Plots...")
    print("Observe the plots for potential skewness and outliers.")

    # Calculate number of rows needed for subplots (2 plots per feature)
    n_features = len(selected_features)

    # Create plots for each selected feature
    for feature in selected_features:
        print(f"\n--- Distribution for: {feature} ---")
        try:
            # Create a figure with two subplots side-by-side
            fig, axes = plt.subplots(1, 2, figsize=(14, 4))
            fig.suptitle(f'Distribution Analysis for Feature: {feature}', fontsize=14)

            # 1. Histogram with KDE (Kernel Density Estimate)
            sns.histplot(data=df, x=feature, kde=True, ax=axes[0], bins=50)
            axes[0].set_title('Histogram with Density Curve')
            axes[0].set_xlabel(f'{feature} Value')
            axes[0].set_ylabel('Frequency')

            # 2. Box Plot
            sns.boxplot(data=df, x=feature, ax=axes[1])
            axes[1].set_title('Box Plot')
            axes[1].set_xlabel(f'{feature} Value')

            plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
            plt.show()

        except Exception as e:
            print(f"An error occurred while plotting feature {feature}: {e}")

    print("\n--- How to Interpret These Plots ---")
    print("1. Histogram:")
    print("   - Look at the overall shape.")
    print("   - A distribution concentrated on one side with a long tail on the other indicates skewness.")
    print("     (Tail to the right = positive skew, Tail to the left = negative skew)")
    print("   - Multiple peaks might suggest different underlying groups.")
    print("2. Box Plot:")
    print("   - The box shows the Interquartile Range (IQR) - the middle 50% of the data.")
    print("   - The line inside the box is the median (50th percentile).")
    print("   - Whiskers typically extend to 1.5 * IQR from the box edges.")
    print("   - Points plotted individually beyond the whiskers are potential outliers.")
    print("   - An off-center median within the box also suggests skewness.")

else:
    print("No selected features were found in the DataFrame to analyze.")
```
