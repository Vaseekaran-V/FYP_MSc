```python
train_dates = ['2017-03-31 00:00:00', '2017-04-30 00:00:00', '2017-05-31 00:00:00',
 '2017-06-30 00:00:00', '2017-07-31 00:00:00', '2017-08-31 00:00:00',
 '2017-09-30 00:00:00', '2017-10-31 00:00:00', '2017-11-30 00:00:00',
 '2017-12-31 00:00:00']#, '2018-01-31 00:00:00', '2018-02-28 00:00:00']

test_dates = [#'2017-04-30 00:00:00', '2017-05-31 00:00:00',
 '2017-06-30 00:00:00', '2017-07-31 00:00:00', '2017-08-31 00:00:00',
 '2017-09-30 00:00:00', '2017-10-31 00:00:00', '2017-11-30 00:00:00',
 '2017-12-31 00:00:00', '2018-01-31 00:00:00', '2018-02-28 00:00:00',
 '2018-03-31 00:00:00']
```

```python
df_train = df[df['end_of_month'].isin(train_dates)]
df_test = df[df['end_of_month'].isin(test_dates)]
```

```python
df_test.end_of_month.value_counts()
```

```python
#Function to return the df with ids for all 7 months
def get_df_ids_all_months(df, customer_id_col = 'customer_ID', date_col = 'end_of_month', num_of_months = 10):
    customer_counts = df.groupby(customer_id_col)[date_col].nunique()
    customer_ids_all_dates = customer_counts[customer_counts == num_of_months].index
    return df[df[customer_id_col].isin(customer_ids_all_dates)]

```

```python
df_train_all_dates = get_df_ids_all_months(df_train)
df_test_all_dates = get_df_ids_all_months(df_test)
```
