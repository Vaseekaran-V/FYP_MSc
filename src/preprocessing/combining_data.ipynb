```python
import pandas as pd
import numpy as np
```

```python
df_last_month = pd.read_parquet('../../data/initial_modelling/train_X_data_last_month_oh.parquet')
df_ts_selected = pd.read_parquet('../../data/ts_features/selected_features/train_13_months_boruta_selected_ts_features.parquet').drop(columns = ['end_of_month'])
df_tsfel_stat_selected = pd.read_parquet('../../data/tsfel_features/selected_features/train_13_months_boruta_selected_tsfel_stat_features.parquet')
df_tsfel_temporal_selected = pd.read_parquet('../../data/tsfel_features/selected_features/train_13_months_boruta_selected_tsfel_temporal_features.parquet')
```

```python
# Drop overlapping columns from the second DataFrame before merging
df_ts_selected = df_ts_selected.drop(columns=[
    col for col in df_last_month.columns if col != 'customer_ID' and col in df_ts_selected.columns
])
df_tsfel_stat_selected = df_tsfel_stat_selected.drop(columns=[
    col for col in df_last_month.columns if col != 'customer_ID' and col in df_tsfel_stat_selected.columns
])
df_tsfel_temporal_selected = df_tsfel_temporal_selected.drop(columns=[
    col for col in df_last_month.columns if col != 'customer_ID' and col in df_tsfel_temporal_selected.columns
])
```

```python
df_tsfel_temporal_selected
```

```python
# Merge all features into a single DataFrame using the 'customer_ID' column
df_merged = df_last_month.merge(
    df_ts_selected, on='customer_ID', how='left'
).merge(
    df_tsfel_stat_selected, on='customer_ID', how='left'
).merge(
    df_tsfel_temporal_selected, on='customer_ID', how='left'
)
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged.isna().sum()
```

```python
df_merged.fillna(0, inplace=True)
df_merged.isna().sum()
```

```python
df_merged.to_parquet('../../data/final_features_ml/train_all_features_last_month.parquet', index=False, engine='pyarrow')
```
