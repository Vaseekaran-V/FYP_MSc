```python
import pandas as pd
import numpy as np
```

```python
train_df_last_month = pd.read_parquet('../../data/initial_modelling/train_X_data_last_month_oh.parquet').sort_values(by='customer_ID').reset_index(drop=True)
train_df_dl_features = pd.read_parquet('../../data/dl_features/exp_25_train_extracted.parquet')
```

```python
train_df_last_month.head()
```

```python
train_df_dl_features.head()
```

```python
train_df_last_month.shape, train_df_dl_features.shape
```

```python
train_df_dl_combined = pd.concat([train_df_last_month, train_df_dl_features], axis=1)
train_df_dl_combined.head()
```

```python
val_df_last_month = pd.read_parquet('../../data/initial_modelling/val_X_data_last_month_oh.parquet').sort_values(by='customer_ID').reset_index(drop=True)
val_df_dl_features = pd.read_parquet('../../data/dl_features/exp_25_val_extracted.parquet')

val_df_dl_combined = pd.concat([val_df_last_month, val_df_dl_features], axis=1)
val_df_dl_combined.head()
```

```python
test_df_last_month = pd.read_parquet('../../data/initial_modelling/test_X_data_last_month_oh.parquet').sort_values(by='customer_ID').reset_index(drop=True)
test_df_dl_features = pd.read_parquet('../../data/dl_features/exp_25_test_extracted.parquet')

test_df_dl_combined = pd.concat([test_df_last_month, test_df_dl_features], axis=1)
test_df_dl_combined.head()
```

```python
train_df_dl_combined.to_parquet('../../data/last_month_and_dl_features/train_X_data_last_month_and_dl_features.parquet', index=False)
val_df_dl_combined.to_parquet('../../data/last_month_and_dl_features/val_X_data_last_month_and_dl_features.parquet', index=False)
test_df_dl_combined.to_parquet('../../data/last_month_and_dl_features/test_X_data_last_month_and_dl_features.parquet', index=False)
```
