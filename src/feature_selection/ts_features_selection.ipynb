```python
df_train.sort_values(by = ['customer_ID'], inplace = True)
y_train.sort_values(by = ['customer_ID'], inplace = True)
```

```python
df_train.isna().sum()
```

```python
#replace null values with zero
df_train.fillna(0, inplace = True)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(random_state=0, tree_method = 'hist', device = 'cuda')
feat_selector = BorutaPy(model, n_estimators='auto', verbose=2, random_state=0)
```

```python
feat_selector.fit(df_train.drop(columns = ['customer_ID', 'end_of_month']), y_train['target'].values)
```

```python
selected_features = pd.DataFrame({'Feature': df_train.drop(columns = ['customer_ID', 'end_of_month']).columns, 'Importance': feat_selector.ranking_})
selected_features.sort_values(by = 'Importance', ascending = True, inplace = True)
```

```python
boruta_features = selected_features[selected_features['Importance'] == 1]['Feature'].to_list()
boruta_features = list(np.append(['customer_ID', 'end_of_month'], boruta_features))
```

```python
selected_features.to_csv('../../utils/selected_features/ts_features_boruta.csv', index = False)
```

```python
df_train[boruta_features].to_parquet('../../data/ts_features/selected_features/train_13_months_boruta_selected_ts_features.parquet', index = False)
```
