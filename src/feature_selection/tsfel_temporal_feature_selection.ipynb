```python
df.sort_values(by = ['customer_ID'], inplace = True)
y_train.sort_values(by = ['customer_ID'], inplace = True)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(random_state=0, tree_method = 'hist', device = 'cuda')
feat_selector = BorutaPy(model, n_estimators='auto', verbose=2, random_state=0)
```

```python
feat_selector.fit(df.drop(columns = ['customer_ID']), y_train['target'].values)
```

```python
selected_features = pd.DataFrame({'Feature': df.drop(columns = ['customer_ID']).columns, 'Importance': feat_selector.ranking_})
selected_features.sort_values(by = 'Importance', ascending = True, inplace = True)
selected_features.head(10)
```

```python
boruta_features = selected_features[selected_features['Importance'] == 1]['Feature'].to_list()
boruta_features = list(np.append(['customer_ID'], boruta_features))
```

```python
selected_features.to_csv('../../utils/selected_features/tsfel_temporal_features_boruta.csv', index = False)
```
